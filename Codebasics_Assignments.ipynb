```python
def is_prime_and_odd(number):
    if number <= 1:
        return False, False  # Neither prime nor odd
    is_prime = True
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False
            break
    is_odd = number % 2 != 0
    return is_prime, is_odd

# Test the function
num = 11
prime, odd = is_prime_and_odd(num)
if prime:
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
if odd:
    print(f"{num} is an odd number.")
else:
    print(f"{num} is not an odd number.")
```

```python
def pay_bill(expenses, percent_commission=None, special_offer_amount=None):
    # Set default commission percentage if not provided
    if percent_commission is None:
        percent_commission = 9.8

    # Calculate total expenses
    total_expenses = sum(expenses)

    # Check if special offer amount is provided
    if special_offer_amount is not None and total_expenses > special_offer_amount:
        # Apply extra commission
        percent_commission += 1.2

    # Calculate commission amount
    commission_amount = total_expenses * (percent_commission / 100)

    # Calculate final payable price
    final_price = total_expenses + commission_amount

    return final_price

# Test the function
expenses = [100, 200, 300]
print("Final payable price:", pay_bill(expenses))
print("Final payable price with special offer:", pay_bill(expenses, special_offer_amount=250))
```

```python
# prompt: Create inheritance using MobilePhone as base class and Apple & Samsung as child class  The base class should have properties:  ScreenType = Touch Screen NetworkType = 4G/5G DualSim = True or False FrontCamera = (5MP/8MP/12MP/16MP) rearCamera = (8MP/12MP/16MP/32MP/48MP) RAM = (2GB/3GB/4GB) Storage = (16GB/32GB/64GB) Create basic mobile phone functionalities in the classes like: make_call, recieve_c

class MobilePhone:
    def __init__(self, screen_type, network_type, dual_sim, front_camera, rear_camera, ram, storage):
        self.screen_type = screen_type
        self.network_type = network_type
        self.dual_sim = dual_sim
        self.front_camera = front_camera
        self.rear_camera = rear_camera
        self.ram = ram
        self.storage = storage

    def make_call(self, number):
        print(f"Calling {number}...")

    def receive_call(self):
        print("Receiving call...")


class Apple(MobilePhone):
    def __init__(self, model, screen_type, network_type, dual_sim, front_camera, rear_camera, ram, storage):
        super().__init__(screen_type, network_type, dual_sim, front_camera, rear_camera, ram, storage)
        self.model = model

    def make_call(self, number):
        print(f"Calling {number} using Apple {self.model}...")


class Samsung(MobilePhone):
    def __init__(self, model, screen_type, network_type, dual_sim, front_camera, rear_camera, ram, storage):
        super().__init__(screen_type, network_type, dual_sim, front_camera, rear_camera, ram, storage)
        self.model = model

    def receive_call(self):
        print(f"Receiving call on Samsung {self.model}...")


apple_iphone_13 = Apple("iPhone 13", "OLED", "5G", True, "12MP", "12MP", "4GB", "128GB")
samsung_galaxy_s22 = Samsung("Galaxy S22", "AMOLED", "5G", True, "10MP", "50MP", "8GB", "256GB")

apple_iphone_13.make_call("0123456789")
samsung_galaxy_s22.receive_call()
```

```python
"""Question3: Group data using custom function:
Here, the requirement is to create four groups based on "imdb_rating"

1 <= imdb_rating <= 3.9 ---> 'Poor'
4 <= imdb_rating <= 7.9 ---> 'Average'
8 <= imdb_rating <= 10 ---> 'Good'
If none of above satisfy, then return 'others'
```

```python
import pandas as pd

# Define custom function to categorize imdb_rating
def categorize_imdb_rating(imdb_rating):
    if 1 <= imdb_rating <= 3.9:
        return 'Poor'
    elif 4 <= imdb_rating <= 7.9:
        return 'Average'
    elif 8 <= imdb_rating <= 10:
        return 'Good'
    else:
        return 'Others'


# Group data using the custom function
grouped_custom = movie.groupby(lambda x: categorize_imdb_rating(movie.loc[x, 'imdb_rating']))

# Print the groups
for key, group in grouped_custom:
    print("Group:", key)
    print(group)
    print()
```

```python
def grouper(movie, idx, col):
    '''
        This function returns category based on imdb_rating
    '''
    if 1 <= movie[col].loc[idx] <= 3.9:
        return 'Poor'
    elif 4 <= movie[col].loc[idx] <= 7.9:
        return 'Average'
    elif 8 <= movie[col].loc[idx] <= 10:
        return 'Good'
    else:
        return "others"
```

```python
#groupby based on the grouper function and store it in a variable 'g'
#Note: use the imdb_rating column
g = movie.groupby(lambda x: grouper(movie, x, 'imdb_rating'))

#iterate through the 'g' and print all the groups
for key, d in g:
    print("Group by Key: {}\n".format(key))
    print(d)
```
