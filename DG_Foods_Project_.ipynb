# Data Exploration

## Dimensionality Reduction with PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components= 2)
pca.fit(dg_data)
```

```python
dg_data_2 = pca.transform(dg_data)

dg_data_2 = pd.DataFrame(dg_data_2)
dg_data_2.index = dg_data.index
dg_data_2.columns = ["PC1", "PC2"]
```

```python
dg_data_2.head(-5)
```

```python
print(pca.explained_variance_ratio_)
```

```python

axis = dg_data_2.plot (kind= "scatter", x= "PC2", y= "PC1", figsize= (18, 9))

for j, Country in enumerate(dg_data.index):
  axis.annotate(Country, (dg_data_2.iloc[j].PC2, dg_data_2.iloc[j].PC1))
```

## Normalization

```python
from sklearn.preprocessing import normalize
```

```python

dg_data_2["Country_mean"] = pd.Series(dg_data.mean(axis= 1), index= dg_data_2.index)
Country_mean_max = dg_data_2["Country_mean"].max()
Country_mean_min = dg_data_2["Country_mean"].min()

Country_mean_scaled = (dg_data_2.Country_mean-Country_mean_min) / Country_mean_max
dg_data_2["Country_mean_scaled"] = pd.Series( Country_mean_scaled, index=dg_data_2.index)
```

```python
dg_data_2.head(-5)
```

## Scattered Plot

```python
dg_data_2.plot(kind= "scatter", x= "PC2", y= "PC1", s= dg_data_2["Country_mean_scaled"]*100, figsize= (18,9))
```

```python
dg_data_2["Country_sum"] = pd.Series(dg_data.sum(axis=1), index= dg_data_2.index)
Country_mean_max = dg_data_2["Country_mean"].max()
Country_mean_min = dg_data_2["Country_mean"].min()

Country_mean_scaled = (dg_data_2.Country_mean-Country_mean_min) / Country_mean_max
dg_data_2["Country_mean_scaled"] = pd.Series( Country_mean_scaled, index=dg_data_2.index)

dg_data_2.plot(kind= "scatter", x= "PC2", y= "PC1", s= dg_data_2["Country_mean_scaled"]*100, figsize= (18,9))
```

```python

dg_data_2["Country_change"] = pd.Series(dg_data["2007"]-dg_data["1990"], index= dg_data_2.index)
Country_change_max = dg_data_2["Country_change"].max()
Country_change_min = dg_data_2["Country_change"].min()

Country_change_scaled = (dg_data_2.Country_change-Country_change_min) / Country_change_max
dg_data_2["Country_change_scaled"] = pd.Series( Country_change_scaled, index=dg_data_2.index)

```

```python
dg_data_2[["Country_change", "Country_change_scaled"]].head(-5)
```

```python
dg_data_2.plot(kind= "scatter", x= "PC2", y= "PC1", s= dg_data_2["Country_mean_scaled"]*100, figsize= (18,9))
```

#K-means Clustering

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans (n_clusters= 5)
clusters = km.fit(dg_data)
```

```python
dg_data_2["clusters"] = pd.Series(clusters.labels_, index=dg_data_2.index)

dg_data_2.head(-5)
```

## Visualization

```python
sns.set_style("whitegrid")

sns.lmplot("PC2", "PC1", data= dg_data_2, hue= "clusters", palette="coolwarm", size= 10, aspect= 1, fit_reg= False)
```

```python
sns.countplot(dg_data_2["clusters"], label= "Count")
plt.show()
```
